## Reproducibility: What is it, What does it mean, Why it's important

The concept of reproducibility is just that \- **is your work reproducible**? Reproduction of methods, analyses and results is important not only in scientific research but also in day\-to\-day activities. It is a major factor that determines if **research results are consistent and valid** using the same methods used in the original study. Many research studies provide significant results but their findings cannot be reproduced. It raises the question of whether the research question or the hypothesis is valid. If your work is reproducible, then **valid experiments and results build authentic discoveries that are more reliable** in your field and help develop robust literature for future research studies. While reproducibility ensures that your work is accurate and reliable, **and encourages community involvement and participation**. _**Remember, if your foundation is shaky then so is all the work built upon it.**_ 

<p align="center"> 
    <img src="./Figures/Reproducibility.jpeg"   width="596.667px"  height="459.661px" /> 
</p>

There are different fields of reproducibility :

1. **Scientific** Reproducibility \- **Can your results be reproduced from your methods?**
2. **Computational** Reproducibility \- **Can your results be reproduced by your code?**
3. **Statistical** Reproducibility \- **Are your methods robust enough to reproduce your results?**

Statistical reproducibility requires **formal training** in study design and interpretation. I.e taking a sample of the population, randomizing treatments, having a large number of replicates, etc – this is part of the methodological reproducibility and statistical reproducibility.

While talk of reproducibility is all good and dandy, it may not be available to everyone. There are significant barriers to reproducibility including:

- **Cost**
- Structure of dissemination does not currently fully support reproducibility
- **Lack of training and incentives**
- **Takes time, held to a higher standard**
- Privacy and security
- Differences between fields

Your job as a scientist and a researcher is to try your best to ensure that your work is reproducible. **If the ability is beyond your means, do not hesitate to say so.** Science is about being open and transparent

There are many tools and conventions to help with your reproducibility journey. These include having a proper workflow and file organization system, as well as GIT for version control

## File Paths

Having a default file structure can also help with your own mental organization of your file paths. There are two types of file paths: **Absolute and relative file paths** 

### Absolute File Paths

Absolute file paths **start at your root folder and mention every directory before arriving at your file**. For example, if I wanted to get the absolute file path of the CSV in the above figure, it would look something like this: `/Users/<username>/Desktop/Projects/Salt Tolerance/Chlamy sample - Sheet1.csv`, Where `Chlamy sample - Sheet1.csv` is the name of my file, which is located in the `Salt Tolerance/` directory. This directory is in the `Projects/` directory which is on my `Desktop/` \(also a directory\), within my user folder \(`<username>`\), within the`Users/` folder in the **root** \(`/`\). The / **denotes a directory \(or also called a folder\)**. Absolute file paths can be very useful, however, **they are very sensitive to breaking**. For example, if I moved `Projects/` to `Documents/` rather than `Desktop/` my path would **no longer work and cause an error**. 

### Relative File Paths

Relative file paths are **relative to the current directory that you are working in**. Your current directory will be noted with a ./ and the parent directory will be denoted with ../ For example if I am in the `Figures/` directory and wish to go to the `References/` directory I can use the relative file path: `../References/.` The benefit of using absolute paths is that if anything happened with `Projects/` (eg: I moved it to `Documents/` from `Desktop/`) **my path would still work**. 

**There is no right or wrong file path format!** **Choose the one that works for you and the type of document that you have.** If you know the hierarchy of your directories, relative paths might serve better. However, if you do not know the hierarchy but know where your file is in general then an absolute path might serve better.

### Configure git global settings on your local computer

Before we start we need to tell git who we are so that we know who is making the files. We can do this using the `git config` commands



```python
%%bash

# Tell git your email. Use your github email
git config --global user.email "<your email>"

# Next tell git your name
git config --global user.name "<username>"

# Change the name of your first branch from "Master" to "main"
git config --global init.defaultBranch main
```

### Hold your horses! The staging environment, the commit and you

Before we get going on our git train, it's important to differentiate between your working directory, the staging environment and a commit. Commits are **records of the changes you've made since your last commit**. These are what allow you to jump between the different stages of your project. However, git does not commit on its own, **you need to tell git what you want to commit**. To do this we first need to **adding the file to the staging area** via the command `git add`. Once you've made a nice package of files in the staging area, you **can then commit your changes to your repo** in the `.git` repository. Let's see what this looks like

<p align="center"> 
    <img src="./Figures/git_staging_area.png"   width="700px"  height="400px" /> 
</p>



```python
%%bash
cd week7_git_lesson

# Now let's add our file to the staging environment
git add first_file.txt

git status # We see now that our file isn't committed but it is about to be
```

```python
%%bash
cd week7_git_lesson

# Now let's commit our file. We have the option to also add a message using the -m flag. It is always beneficial to write something useful so that when you are looking back between your versions it is easy to find what you're looking for 
git commit -m "Created empty file called first_file.txt"

git status # We see now that our file isn't committed but it is about to be
```

```python
%%bash
cd week7_git_lesson

# Great! You just created your first commit! Let's see what happens when we edit our file to add some information to it!
vim first_file.txt


git status #see that the file has been modified

# what was different from the last commit?
git diff first_file.txt

git add first_file.txt #stage our changes
git status 

git commit -m "Added some filler text to file" 
git status
```

### Branching in Git

Branching is a great way to work on a new idea, bug fixes or developing features **without affecting your main/master branch**. They are designed to be **temporary** and to facilitate edits and trials without breaking your original codebase. They allow you to keep your original `main` branch as is. For example, say I had a website and I wanted to make a new webpage. I can branch off of my main branch, create my webpage and then merge once it is completed. This way my website still stays functional as I work on my new webpage. 

<p align="center"> 
    <img src="./Figures/branches.png"   width="700px"  height="400px" /> 
</p>

There are two phases of branch creation: **1\) Making the branch and 2\) Moving to that branch**. Both of these steps can be combined into one but let us take it one at a time. First, let's make a new branch called new\_files. Just like commits and variable names, it helps to have meaningful branch names so that you know what that branch is meant to be working on 


```python
%%bash
cd week7_git_lesson

git branch new_files
git log
git status # notice that while we made a branch our status has not changed
git branch --list

# move to that branch
git checkout new_files
git branch --list

git log
git status # now our branch has changed
```

```python
%%bash
cd week7_git_lesson

# Let's make some changes to our files

# edit first_file.txt

cp first_file.txt second_file.txt # make a copy called second_file.txt
git status

# edit the second_file.txt

git commit -a  -m "Added another line of text about the weather to file" #Notice how it only added changes and committed for the first_file.txt as git was watching it

# Let's add and commit the other files
git add *
git commit -m "adding second file"

# now let's merge our branch back to the main one
# First head back to the main branch
git checkout main
ls

git branch --list

# Merge your branch
git merge new_files

# Delete the old branch since there's no use for it anymore 
git branch -d new_files

git status
```

### Going back to a previous commit

Sometimes we had versions of our code that worked better or documents that were worded more eloquently, but we no longer have them as we've modified the same file. With Git, **as long as the file was committed** we can revert to that version. Let's play around with it:


```python
%%bash
cd week7_git_lesson

#make a temporary file and commit it
touch temp.txt
git add temp.txt
git commit -m "adding temporary file temp.txt to delete"

# Remove the file and update git
rm temp.txt
git status # while we deleted the file locally, it is still in our repo
git rm temp.txt
git commit -m "deleted temporary file temp.txt"

#make some minor changes
echo temp2 > temp2.txt
git add temp2.txt 
git commit -m "added temp2.txt"


# Revert back to the previous commit
# get the commit ID
git log
git revert <commit_id>
ls
git commit -m "reverting back to other commit with temp.txt"

# Revert but keep changes, can also do with the one above ^ however this one will delete the commits following the one you want to go to
touch temp3.txt
git add temp3.txt
git commit -m "added temp3.txt"
rm temp3.txt 
git restore temp3.txt
git commit -m "message"

# Other options
git checkout <commit_id>
git checkout main # return to original

git reset --hard <commit_id>
```

## Collaborative Repos

Great! Now that we've pushed files to GitHub let's try to use it in a collaborative way. There are two common collaborative workflows:

1. The Shared Repository Workflow
   - Assigned collaborators work on their own branch and then merge into the main
2. The Fork and Pull Workflow
   - Collaborators fork the repository (make a personal copy) and work on their own fork. Merging requires pull requests. The project lead will accept pull requests \(PRs\), review and accept changes made

Let's see how the shared repo workflow looks!

### Shared Repo

We did a lot of Git today, and we know there is a lot of jargon surrounding git so here is a brief recap:

| Term      | Description |
| ----------- | ----------- |
| Add | Adding your changes to the “Staging” area|
| Branch | A parallel version of a repository where changes can be made that do not affect the main repository. New branches are often used to test changes or new ideas, which can later be merged with the base branch|
| Checkout | Moving from one branch to another|
| Clone | Make a copy of the your GitHub repo on your local computer|
| Commit | Takes the files in your staging area and store their changes permanently in your Git directory|
| Head | Head is a reference or pointer of the latest commit on your branch, there can only be one head at a time. You can move the head to a different commit by using checkout - this is known as a detatched head |
| Merge | Incorporate changes into the branch you are on|
| Pull | A combination of fetch and merge. This fetches the files from the remote repo and merges them with your local files|
| Pull Request | Term used in collaboration. You “issue a pull request” to the owner of the upstream repo asking them to pull your changes into their repo (accept your work)|
| Push | Sync your file changes to the remote repo on GitHub|
